### First, import pandas library

```python
import pandas as pd
```

### Read in the table.  We want the second table on that page.

```python
df = pd.read_html("http://www.gurobi.com/resources/examples/food-manufacture-II")[1]
df
```

### Set the index to be the Month column

```python
df1 = df.set_index(['Month'])
df1
```

### Reset the index

```python
df2 = df1.reset_index()
df2
```

### Take the DataFrame that has an index, reset it, and set a new index (to avoid losing the original index column)

```python
df3 = df1.reset_index().set_index(['VEG 1'])
df3
```
